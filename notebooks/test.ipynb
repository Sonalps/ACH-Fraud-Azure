```python
import pandas as pd
import json
import requests

df = pd.read_csv("../data/creditcard_features.csv")
row = df.iloc[0]

# Loading feature list
import joblib
feature_cols = joblib.load("models/feature_columns.pkl")

# Building the payload dict with only the features the API expects
features = {col: float(row[col]) for col in feature_cols if col in df.columns}

payload = {
    "features": features
}

print("Sample payload:\n", json.dumps(payload, indent=2)[:800])  
```

```python
response = requests.post(
    "http://localhost:8000/score",
    json=payload
)

print(response.status_code, response.json())
```
